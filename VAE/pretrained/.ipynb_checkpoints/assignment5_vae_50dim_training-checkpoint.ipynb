## Variational Auto-Encoder (VAE) 

Variational auto-encoder (abbrv. VAE, Kingma and Welling - 2013) is a generative model. Like an auto-encoder, the VAE projects input data into a latent space and then reconstructs it back. But instead of mapping the data to a single point in the embedding space (now we will refer to this as the *latent space*) it maps it to a distribution. 

In the VAE we map each datapoint to a mean and standard deviation of a multivariate Gaussian (the number of dimensions of this Gaussian is the dimension of our latent space). We then try to 'overlap' a standard Gaussian (zero mean and unit variance, with same dimensions as the latent space) over this distribution of datapoints. This is what minimizing the KL-Divergence achieves.

Now, to get a re-construction, we just sample from the standard gaussian and project the sample back into the original data space (To the space of MNIST images, in our case). We also, optimize the parameters of the model to maximize the 'reconstruction' or minimize the 'difference' between the samples and the input image.

So, in the VAE we have a sum of two terms in the cost function:
1. The KL-Divergence term (the 'overlap' term discussed above).
2. The Reconstruction term (also referred to as the 'reconstruction' term above).

Contrast this with the auto-encoder which just has the 'reconstruction' as the cost function.

### Drawing samples from the VAE

Recall that the VAE is a generative model, so we must be able to draw samples from it. Minimizing the 'overlap' term in the VAE cost allows us to approximate a complex multivariate normal (which can be difficult/impossible to sample from) to a standard multivariate normal (easy to sample from). After training the VAE, we should then be able to sample from the standard multivariate normal and project them to data space to get samples of (in this case) MNIST images. So, we have a generative model !

Now that we have an intuitive understanding of the VAE, let's implement it.

```python
datasets = load_data('mnist.pkl.gz', train_size = 50000,binarize="stochastic")
train_x, train_y = datasets[0]
valid_x, valid_y = datasets[1]
test_x, test_y = datasets[2]
```

```python
def gradient_updates_Adam(cost, params, learning_rate,eps = 1e-8,beta1 = 0.9,beta2 = 0.999):
    # Function to return an update list for the parameters to be updated
    
    # cost: MSE cost Theano variable
    # params :  parameters coming from hidden and output layers
    # learning rate: learning rate defined as hyperparameter
    
    # Outputs:
    # updates : updates to be made and to be defined in the train_model function. 
    updates = []
    for param in params:
             
            t = theano.shared(1)
            s = theano.shared(param.get_value(borrow=True)*0.)
            r = theano.shared(param.get_value(borrow=True)*0.)
            updates.append((s, beta1*s + (1.0-beta1)*T.grad(cost, param) ))
            updates.append((r, beta2*r + (1.0-beta2)*(T.grad(cost, param)**2) ))
            s_hat =  s/(1-beta1**t)
            r_hat =  r/(1-beta2**t)
            updates.append((param, param - learning_rate*s_hat/(np.sqrt(r_hat)+eps) ))
            updates.append((t, t+1)) 

    return updates
```

```python
# implementing the VAE in theano
# similar network architecture to (Kingma and Welling 2013)
import numpy as np
import theano
import theano.tensor as T
from theano.sandbox.rng_mrg import MRG_RandomStreams as RandomStreams

input = T.fmatrix("input") # symbolic float32 matrix, eventually will have (minibatch x 784)
learning_rate = T.fscalar("learning_rate") # float32 learning rate.
mbidxs = T.lvector("mbidxs")

# autoencoder layer dimensions
latent_dim = 50
dims = [784, 500, latent_dim, 500, 784]

uniform_lim = lambda n_in,n_out : (-np.sqrt(6.0/(n_in+n_out)), np.sqrt(6.0/(n_in+n_out))) 

W1 = theano.shared(rng_np.uniform(*uniform_lim(dims[0], dims[1]), size=dims[0:2]), borrow=True, name="W1")
W2_mu = theano.shared(rng_np.uniform(*uniform_lim(dims[1], dims[2]), size=dims[1:3]), borrow=True, name="W2_mu")
W2_logvar = theano.shared(rng_np.uniform(*uniform_lim(dims[1], dims[2]), size=dims[1:3]), borrow=True, name="W2_logvar")

W3 = theano.shared(rng_np.uniform(*uniform_lim(dims[2], dims[3]), size=dims[2:4]), borrow=True, name="W3")
W4 = theano.shared(rng_np.uniform(*uniform_lim(dims[3], dims[4]), size=dims[3:5]), borrow=True, name="W4")

b1 = theano.shared(np.zeros((dims[1],)), borrow=True, name="b1")
b2_mu = theano.shared(np.zeros((dims[2],)), borrow=True, name="b2_mu")
b2_logvar = theano.shared(np.zeros((dims[2],)), borrow=True, name="b2_logvar")

b3 = theano.shared(np.zeros((dims[3],)), borrow=True, name="b3")
b4 = theano.shared(np.zeros((dims[4],)), borrow=True, name="b4")

relu = T.nnet.relu
sigmoid = T.nnet.sigmoid
tanh = T.tanh

srng = RandomStreams(seed=1234)
x1     = tanh(T.dot(input, W1) + b1)
mu     = T.dot(x1, W2_mu) + b2_mu
logvar = T.dot(x1, W2_logvar) + b2_logvar
z = mu + srng.normal(logvar.shape)*T.exp(0.5*logvar)
x1_rec = tanh(T.dot(z, W3) + b3)
x_rec  = sigmoid(T.dot(x1_rec, W4) + b4)

cost_rec = -T.nnet.binary_crossentropy(x_rec, input).sum(axis=1)
cost_KL   = 0.5*T.sum(1 + logvar - mu**2 - T.exp(logvar), axis=1)
lower_bound = T.mean(cost_KL + cost_rec)
cost = -lower_bound # minimize -ve of lower bound

params = [W1,W2_mu,W2_logvar,W3,W4,b1,b2_mu,b2_logvar,b3,b4]
updates = gradient_updates_Adam(cost,params,learning_rate)


vae_train = theano.function(
        inputs = [mbidxs, learning_rate], 
        outputs = [cost, x_rec, mu, logvar, cost_KL, cost_rec],
        updates = updates,
        givens = {
            input : train_x[mbidxs,:],
        }
    )


vae_predict = theano.function(
        inputs = [input],
        outputs = [cost, x_rec, mu, logvar]
    )


vae_generate = theano.function(
        inputs = [z],
        outputs = x_rec
    )
```

```python
mb_size = 500
epochs = 500
lrate = 0.0001
train_errors = []
valid_errors = []
train_rec = []
train_kl = []
train_start_time = time.time()

train_size = len(train_y.eval()) 
training_idxs = np.arange(train_size)

print("***** training started  *****")

for epoch in range(epochs):
    
    rng_np.shuffle(training_idxs)
                
    try:    
        t_minib_errors = []
        t_minib_cost_kl = []
        t_minib_cost_rec = []
        start_time = time.time()
        
        for batch_startidx in range(0, train_size, mb_size):

            mini_batch = training_idxs[batch_startidx:batch_startidx+mb_size]
            
            rec_error, rec, train_mu, train_logvar, cost_kl, cost_rec  = vae_train(
                    mbidxs = np.asarray(mini_batch).astype(np.int32),
                    learning_rate=lrate)
            
            t_minib_errors.append(rec_error)
            t_minib_cost_kl.append(cost_kl)
            t_minib_cost_rec.append(cost_rec)
            
        valid_rec_error, valid_rec, valid_mu, valid_logvar = vae_predict(input = valid_x.get_value())
        
        
        train_errors.append(np.mean(t_minib_errors))
        train_rec.append(np.mean(t_minib_cost_rec))
        train_kl.append(np.mean(t_minib_cost_kl))
        valid_errors.append(np.mean(valid_rec_error))
        print("epoch={:<3d} -- train_error={:>6.4f} -- valid_error={:>6.4f} -- cost(Rec,KL)=({:>6.4f},{:>6.4f}) -- time={:>5.3f}".format(epoch, train_errors[-1], valid_errors[-1], np.mean(t_minib_cost_rec),np.mean(t_minib_cost_kl),time.time()-start_time))    
    
    
    except KeyboardInterrupt as e:
        print("***** stopping *****")
        break
```
